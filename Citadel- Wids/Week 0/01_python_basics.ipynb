### **Exercise 1: Load Data and Initial Inspection**

**Objective:** Load the `AAPL.csv` dataset into a pandas DataFrame and perform an initial inspection to understand its structure and data types. This step is crucial for preparing the data for further analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the AAPL.csv dataset
df = pd.read_csv('AAPL.csv')

# Display the first few rows to preview the data
display(df.head())

# Display data types and non-null values to check for any issues
df.info()
```

### **Exercise 2: Data Preprocessing and Feature Engineering**

**Objective:** Convert the 'Date' column to a datetime object and set it as the DataFrame's index. Then, calculate the daily percentage returns of the 'Close' price and a 20-day Simple Moving Average (SMA). These are common metrics used in financial analysis.

```python
# Convert 'Date' column to datetime objects
#df['Date'] = pd.to_datetime(df['Date'])

# Set the 'Date' column as the DataFrame's index
#df.set_index('Date', inplace=True)
#df['Close/Last'] = df['Close/Last'].str.replace('$', '', regex=False).astype(float)
# Calculate daily returns using the percentage change of the 'Close' price
df['Daily Returns'] = df['Close/Last'].pct_change()

# Calculate the 20-day Simple Moving Average (SMA) of the 'Close' price
df['20-Day SMA'] = df['Close/Last'].rolling(window=20).mean()

# Display the first few rows of the DataFrame with the new columns
display(df.head(40))
```

### **Exercise 3: Visualize Closing Price and Moving Average**

**Objective:** Plot the historical 'Close' prices alongside the calculated '20-Day SMA'. This visualization helps in understanding the trend of the stock price and how the moving average smooths out price fluctuations, often indicating potential support or resistance levels.

```python
# Create a figure and axes for the plot
plt.figure(figsize=(14, 7))

# Plot the 'Close' price
plt.plot(df['Close/Last'], label='Closing Price', color='blue')

# Plot the '20-Day SMA'
plt.plot(df['20-Day SMA'], label='20-Day SMA', color='red', linestyle='--')

# Add a title and labels for clarity
plt.title('AAPL Closing Price and 20-Day Simple Moving Average')
plt.xlabel('Date')
plt.ylabel('Price (USD)')

# Add a legend to distinguish between the plotted lines
plt.legend()

# Add a grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```
